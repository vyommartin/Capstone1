Undoubtedly, I am quite frustrated by how the accuracy is stagnating but what I also seem to notice from this is that there might be two situations:

1.   We are stuck in **Local Minimum**.
2.   We are bouncing off from the same two points on the either 
sides of the **Global Minimum**.

So what can we do about?

Lets assume that the loss our model is bouncing off the same two points in gradient descent and thus we must lower the learning rate.
I will introduce some weight decay too so that it slows down when and if it reaches the global minimum. If this does not work, put on the seat belts because I will introduce momentum and a lower learning rate to make sure we are out of the local minimum valley!

```
learner.fit_one_cycle(5, 1e-5, wd=0.05)
```

```
learner.save('94%')
```

```
learner.load('94%')
```

```
learner.fit_one_cycle(1, 1e-5, wd=0.05)
```
